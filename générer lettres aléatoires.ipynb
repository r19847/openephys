```python
import random
import matplotlib.pyplot as plt

# Couleurs
ROUGE_CLAIR = '#FF8080'
BLEU_CLAIR = '#8080FF'

# Gnrer les essais avec les bras gauche (G) ou droit (D) tout en vitant plus de 3 lettres conscutives identiques
def generate_trials():
    essais = []
    while len(essais) < 60:
        choix = random.choice(['G', 'D'])
        if len(essais) >= 3 and essais[-1] == essais[-2] == essais[-3] == choix:
            continue
        essais.append(choix)
    return essais

# Gnrer les rsultats des essais
def generate_results():
    return [''] * 60

# Crer une figure
fig, axs = plt.subplots(7, figsize=(8.27, 11.69))  # Taille A4 (en pouces)

# Dfinir le titre
plt.suptitle('Rate 2', fontsize=40, fontweight='bold')

# Gnrer et afficher les tableaux
for ax in axs:
    essais = generate_trials()
    resultats = generate_results()

    ax.axis('off')
    table = ax.table(cellText=[[f"{i+1}" for i in range(60)], essais, resultats, [''] * 60],  # Ajout d'une ligne vide supplmentaire
                    colWidths=[0.05] * 60,
                    cellLoc='center',
                    loc='center')

    # Coloration des cases et mise en gras des lettres
    for i in range(len(essais)):
        if essais[i] == 'G':
            table.get_celld()[1, i].set_facecolor(ROUGE_CLAIR)
            table.get_celld()[1, i].set_text_props(weight='bold')
        elif essais[i] == 'D':
            table.get_celld()[1, i].set_facecolor(BLEU_CLAIR)
            table.get_celld()[1, i].set_text_props(weight='bold')

    table.auto_set_font_size(False)
    table.set_fontsize(8)

plt.show()
```
